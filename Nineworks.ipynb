### 2번

```python
cursor.execute("SELECT DISTINCT c.concept_name FROM concept AS c \
                JOIN condition_occurrence AS co \
                ON co.condition_concept_id = c.concept_id \
                WHERE c.concept_name LIKE '%heart%' AND (\
               (c.concept_name LIKE 'a%') OR (c.concept_name LIKE 'A%') OR \
               (c.concept_name LIKE 'b%') OR (c.concept_name LIKE 'B%') OR \
               (c.concept_name LIKE 'c%') OR (c.concept_name LIKE 'C%') OR \
               (c.concept_name LIKE 'd%') OR (c.concept_name LIKE 'D%') OR \
               (c.concept_name LIKE 'e%') OR (c.concept_name LIKE 'E%'));")
cursor.fetchall()
```

### 3번

```python
cursor.execute("SELECT drug_concept_id, drug_exposure_start_date,drug_exposure_end_date  \
                FROM drug_exposure WHERE person_id = '1891866' GROUP BY drug_concept_id;")
cursor.fetchall()
```

```python
'''
1.datetime 형식을 바꾸는 방법
2.datetime을 string으로 인식 후 y-m-d 형식으로 변환
3.Group by drug_concept_id 로 그룹화
4.Order (By 새롭게 계산한 칼럼 = 종료일 - 시작일) DESC
'''
```
